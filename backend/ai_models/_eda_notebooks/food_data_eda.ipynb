# Exploratory Data Analysis (EDA) for Food Data

This notebook analyzes the `FOOD-DATA.csv` dataset, which is used by the `NutritionDatabase` model. The goal is to explore the distribution of nutrients and their correlations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the dataset
csv_path = os.path.join('..', 'food_data', 'Data', 'FOOD-DATA.csv')
df = pd.read_csv(csv_path)

print('Dataset loaded successfully.')
```

## 1. Data Overview

Let's start by examining the first few rows of the dataset and getting a summary of its contents.

```python
print('First 5 rows of the dataset:')
display(df.head())

print('\nDataset Info:')
df.info()
```

## 2. Nutrient Distributions

Now, let's visualize the distributions of some key nutrients.

```python
key_nutrients = ['Protein', 'Fat', 'Carbohydrates', 'Sugars', 'Dietary Fiber']

plt.figure(figsize=(15, 10))
for i, nutrient in enumerate(key_nutrients, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[nutrient], kde=True)
    plt.title(f'Distribution of {nutrient}')
plt.tight_layout()
plt.show()
```

## 3. Nutrient Correlations

Let's explore the correlations between these key nutrients.

```python
correlation_matrix = df[key_nutrients].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Key Nutrients')
plt.show()
```
