# Exploratory Data Analysis (EDA) for Allergy Analyzer

This notebook analyzes the `allergies_processed.csv` dataset, which is used by the `AllergyAnalyzer` model. The goal is to understand the distribution and relationships of allergens in the food data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the dataset
csv_path = os.path.join('..', 'allergy_analyzer', 'allergies_processed.csv')
df = pd.read_csv(csv_path)

print('Dataset loaded successfully.')
```

## 1. Data Overview

Let's start by examining the first few rows of the dataset and getting a summary of its contents.

```python
print('First 5 rows of the dataset:')
display(df.head())

print('\nDataset Info:')
df.info()
```

## 2. Allergen Distribution

Now, let's visualize the distribution of allergens across the dataset. This will show us which allergens are most common.

```python
allergen_cols = df.columns[1:]
allergen_counts = df[allergen_cols].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=allergen_counts.values, y=allergen_counts.index, palette='viridis')
plt.title('Distribution of Allergens')
plt.xlabel('Number of Foods')
plt.ylabel('Allergen')
plt.show()
```
